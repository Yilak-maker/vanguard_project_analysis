DURATION IN STEPS:

```python
# calculating the time process
df_final['date_time'] = pd.to_datetime(df_final['date_time'])
# Sort data by visitor_id, visit_id, and date_time
df_final = df_final.sort_values(by=['visitor_id', 'visit_id', 'date_time'])
# Calculate time difference to next step within each visitor & visit group
df_final['next_date_time'] = df_final.groupby(['visitor_id', 'visit_id'])['date_time'].shift(-1)
# Calculate duration spent on each step (in seconds or minutes)
df_final['time_spent_sec'] = (df_final['next_date_time'] - df_final['date_time']).dt.total_seconds()
# Optional: Convert seconds to minutes
df_final['time_spent_min'] = df_final['time_spent_sec'] / 60

#aggregate average time spent per nstep and per group
avg_time_by_step_variation = df_final.groupby(['Variation', 'process_step'])['time_spent_min'].mean().reset_index()
print(avg_time_by_step_variation)

pivot_table = avg_time_by_step_variation.pivot(index='process_step', columns='Variation', values='time_spent_min')
print(pivot_table)

# Your pivot table: index=process_step, columns=Variation, values=time_spent_min
pivot_table = avg_time_by_step_variation.pivot(index='process_step', columns='Variation', values='time_spent_min')
# Calculate the total time per group (sum over steps)
total_time_per_group = pivot_table.sum(axis=0)
# Calculate percentage per step per group
percentage_table = pivot_table.div(total_time_per_group, axis=1) * 100
print(percentage_table)

```

ERROR RATES:

```python
step_order = {"start": 0, "step_1": 1, "step_2": 2, "step_3": 3, "confirm": 4}
df_final['step_num'] = df_final['process_step'].map(step_order)
```

```python
df_final = df_final.sort_values(by=['visit_id', 'date_time'])
```

```python
results = []

for variation in df_final['Variation'].unique():
    df_var = df_final[df_final['Variation'] == variation].copy()
    # Get previous step number for each visit
    df_var['prev_step_num'] = df_var.groupby('visit_id')['step_num'].shift(1)

    # Check if current step is less than previous step (regression)
    df_var['regression'] = df_var['step_num'] < df_var['prev_step_num']
    # Group by visit and check if any step regressed
    regression_summary = df_var.groupby('visit_id')['regression'].any().reset_index()

    # Convert True/False to 1/0
    regression_summary['regression'] = regression_summary['regression'].astype(int)

    total_visits = regression_summary.shape[0]
    visits_with_regression = regression_summary['regression'].sum()
    error_rate = (visits_with_regression / total_visits) * 100

    results.append({
        'Variation': variation,
        'Total Visits': total_visits,
        'Visits with Regression': visits_with_regression,
        'Error Rate (%)': round(error_rate, 2)
    })

```

```python
results_df = pd.DataFrame(results)
print(results_df)
```

```python
df_final['prev_step_num'] = df_final.groupby('visit_id')['step_num'].shift(1)
df_final['regression'] = df_final['step_num'] < df_final['prev_step_num']

# Step 2: Filter only rows where regression occurred
regression_steps = df_final[df_final['regression']].copy()

# Step 3: Count regressions by step
regression_counts = regression_steps['step_num'].value_counts().rename('regression_count')

# Step 4: Count total visits to each step
total_counts = df_final['step_num'].value_counts().rename('total_count')

# Step 5: Combine and compute error rate
step_stats = pd.concat([regression_counts, total_counts], axis=1).fillna(0)
step_stats['error_rate (%)'] = (step_stats['regression_count'] / step_stats['total_count']) * 100
step_stats = step_stats.sort_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=step_stats.index, y=step_stats['error_rate (%)'])
plt.title("Error Rate by Step")
plt.xlabel("Step Number")
plt.ylabel("Error Rate (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Add previous step and regression column
df_final['prev_step_num'] = df_final.groupby('visit_id')['step_num'].shift(1)
df_final['regression'] = df_final['step_num'] < df_final['prev_step_num']

# Step 2: Total step visits by variation and step
total_counts = df_final.groupby(['Variation', 'step_num']).size().reset_index(name='total_count')

# Step 3: Regressions by variation and step
regression_counts = df_final[df_final['regression']].groupby(['Variation', 'step_num']).size().reset_index(name='regression_count')

# Step 4: Merge and calculate error rate
step_variation_stats = pd.merge(total_counts, regression_counts, on=['Variation', 'step_num'], how='left')
step_variation_stats['regression_count'] = step_variation_stats['regression_count'].fillna(0)
step_variation_stats['error_rate (%)'] = (step_variation_stats['regression_count'] / step_variation_stats['total_count']) * 100
step_variation_stats = step_variation_stats.sort_values(by=['Variation', 'step_num'])
```

```python
#to avoid NaNs
drop_off = step_counts.copy()
for i in range(1, len(step_counts.columns)):
    current_step = step_counts.iloc[:, i]
    previous_step = step_counts.iloc[:, i - 1]
    with np.errstate(divide='ignore', invalid='ignore'):
        drop_off.iloc[:, i] = 100 * (1 - (current_step / previous_step))
        drop_off.iloc[:, i] = drop_off.iloc[:, i].replace([np.inf, -np.inf], np.nan).fillna(0)
drop_off = drop_off.round(2)
print(drop_off)
```

```python
#drop off by age 
step_counts_age = df_final.groupby(['Variation', 'client_age', 'process_step'])['client_id'].nunique().unstack().fillna(0)

drop_off_age = step_counts_age.copy()
for i in range(1, len(step_counts_age.columns)):
    current = step_counts_age.iloc[:, i]
    previous = step_counts_age.iloc[:, i-1]
    drop_off_age.iloc[:, i] = 100 * (1 - (current / previous))

drop_off_age = drop_off_age.round(2)
print(drop_off_age)
```

HYPOTHESIS TESTING

HYPOTHESIS ONE

```python
# If the completion rate in control group = test group.It is H0
# If the completion rate in control group != test group.It is H1
# P test
```

```python
#Filter to final step
final_step = 'confirm'  
df_confirm = df_final[df_final['process_step'] == final_step]

completion_counts = df_confirm.groupby('Variation')['client_id'].nunique()
print("Completions per group:", completion_counts)

total_counts = df_final[df_final['process_step'] == df_final['process_step'].min()].groupby('Variation')['client_id'].nunique()
print("Total clients per group:", total_counts)
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Successes (completions) and total for each group
successes = [completion_counts['Control'], completion_counts['Test']]
totals = [total_counts['Control'], total_counts['Test']]

# Two-sided test
stat, p_value = proportions_ztest(count=successes, nobs=totals)

print(f"Z-test statistic: {stat:.3f}")
print(f"P-value: {p_value:.5f}")

alpha = 0.05  

if p_value < alpha:
    print("❗️Reject the null hypothesis — the completion rates are significantly different.")
else:
    print("✅Fail to reject the null hypothesis — no significant difference in completion rates.")
```

HYPOTHESIS TWO

```python
# The average duration spent in control group = test group. It is H0
# The average duration spent in control group != test group. It is H1
# T test-independent test
```

```python
df_final['date_time'] = pd.to_datetime(df_final['date_time'])

visit_durations = (
    df_final.groupby(['visit_id', 'Variation'])['date_time']
    .agg(duration=lambda x: (x.max() - x.min()).total_seconds()/60)
    .reset_index()
)
```

```python
visit_durations
```

```python
import scipy.stats as st
from scipy.stats import ttest_ind
```

```python
type(visit_durations)
```

```python
control = visit_durations[visit_durations['Variation'] == 'Control']['duration']
test = visit_durations[visit_durations['Variation'] == 'Test']['duration']

t_stat,p_value = ttest_ind(control,test,equal_var=False)

print("T-statistics:",t_stat)
print("p-value:",p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant difference in average durations.")
else:
    print("Fail to reject H0: No significant difference in average durations.")
```
